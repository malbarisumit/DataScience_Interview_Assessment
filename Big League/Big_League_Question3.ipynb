## Calculating away team's offensive efficiency for each game

```python
result['off_e_away'] = result.apply(lambda x: (x.AwayScore)/(x.AwayFGA - x.AwayORB + x.AwayTOV + (.475*x.AwayFTA)),axis=1)
```

```python
result['off_e_away'].head()
```

## Calculating away team's defensive efficiency for each game

```python
result['def_e_away'] = result.apply(lambda x:(x.HomeScore)/(x.HomeFGA - x.HomeORB + x.HomeTOV + (.475*x.HomeFTA)),axis=1)
```

```python
result['def_e_away'].head()
```

## Net away efficiency for each game

```python
result['net_away_efficiency'] = result['off_e_away'] - result['def_e_away']
```

```python
result['net_away_efficiency']
```

## Calculating home team's offensive efficiency for each game

```python
result['off_e_home'] = result.apply(lambda x: (x.HomeScore)/(x.HomeFGA - x.HomeORB + x.HomeTOV + (.475*x.HomeFTA)),axis=1)
```

```python
result['off_e_away'].head()
```

## Calculating home team's defensive efficiency for each game

```python
result['def_e_home'] = result.apply(lambda x:(x.AwayScore)/(x.AwayFGA - x.AwayORB + x.AwayTOV + (.475*x.AwayFTA)),axis=1)
```

```python
result['def_e_home'].head()
```

## Net Home efficiency for each game

```python
result['net_home_efficiency'] = result['off_e_home'] - result['def_e_home']
```

```python
result['net_home_efficiency']
```

```python
team_eff = result[['AwayTeam','net_away_efficiency','HomeTeam','net_home_efficiency']]
```

```python
team_eff
```

```python
away_eff=team_eff.groupby(['AwayTeam']).sum()
away_eff=away_eff.drop(['net_home_efficiency'],axis=1)
```

## Overall net efficiency of Away team

```python
away_eff
```

```python
home_eff=team_eff.groupby(['HomeTeam']).sum()
home_eff=home_eff.drop(['net_away_efficiency'],axis=1)
```

## Overall net efficiency of Home Team

```python
home_eff
```

```python
home_eff.reset_index(inplace=True)
away_eff.reset_index(inplace=True)
```

```python
home_eff.head()
```

## Overall efficiency of a team

```python
overall = pd.concat([home_eff,away_eff],axis=1)
```

```python
overall['Total_Efficiency'] = overall['net_home_efficiency'] + overall['net_away_efficiency']
```

```python
overall=overall.drop(columns=['net_home_efficiency','net_away_efficiency','AwayTeam'],axis=1)
```

```python
final = overall.sort_values(by=['Total_Efficiency'],ascending=False)
```

```python
final['Rank'] = final['Total_Efficiency'].rank(method='dense', ascending=False)
```

## NBA Draft Lottery 

Plackett - Luce model

Import required libraries

```python
import pandas as pd
import numpy
from numpy import *
```

Initial probabilites(Hypothetical)

```python
initialp = [0.114,0.113,0.112,.111,.099,.089,.079,.069,.059,.049,.039,.029,.019,.009,.006,.004]
```

```python
def find(probs,rem_probs,rest_p):
    for i in range(11):
        #print(probs)
        nonzeros = [j for j,e in enumerate(probs) if e!=0]
        team = min(nonzeros)
        #print(team)
        rem_probs[i][team] = rem_probs[i][team]+rest_p
        #print(rem_probs)
        probs[team] = 0
        
    return rem_probs
```

```python
secondp = [0.]*16
thirdp = [0.]*16
fourthp = [0.]*16
fifthp = [0.]*16
#r_probs = [[0 for i in range(16)] for j in range(11)]
r_probs = [[0]*16 for _ in range(11)]
```

```python
for i in range(16):
    first_removed = initialp[:i] + [0.] + initialp[i+1:]
    #print(first_removed)
    total_prob = sum(first_removed)
    c_p = [initialp[i]*x/total_prob for x in first_removed]
    #print(c_p[0])
    secondp = [float("{0:.5f}".format(x+y)) for x,y in zip(secondp,c_p)]
    
    for j in range(16):
        if i!=j:
            second = first_removed[:j]+[0.]+first_removed[j+1:]
            total = sum(second)
            cp_2 = [c_p[j]*x/total for x in second]
            thirdp = [float("{0:.5f}".format(x+y)) for x,y in zip(thirdp,cp_2)]
            
            for k in range(16):
                if i!=k and j!=k:
                    third = second[:k]+[0.]+second[k+1:]
                    total_1 = sum(third)
                    cp_3 = [cp_2[k]*x/total_1 for x in third]
                    fourthp = [float("{0:.5f}".format(x+y)) for x,y in zip(fourthp,cp_3)]
                    
                    for l in range(16):
                        if i!= l and j!=l and k!=l:
                            fourth = third[:l]+[0.]+third[l+1:]
                            total_2 = sum(fourth)
                            cp_4 = [cp_3[l]*x/total_2 for x in fourth]
                            fifthp = [x+y for x,y in zip(fifthp,cp_4)]
                    
                            for m in range(16):
                                if i!=m and j!=m and k!=m and l!=m:
                                #if len(unique([i,j,k,l,m])) == 5:
                                        #print(cp_4[l])
                                        restProb = cp_4[m]
                                        #print(restProb)
                                        f_removed = fourth[:m]+[0.]+fourth[m+1:]
                                        #print(f_removed)
                                        r_probs = find(f_removed,r_probs,restProb)
```
